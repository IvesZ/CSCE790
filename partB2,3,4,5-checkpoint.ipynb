## Example1

```python
import numpy as np
import math
```

```python
x=range(0,500)
```

```python
uk=[]
for i in x:
    u=(math.sin(2*i*math.pi/250)+math.sin(2*i*math.pi/25))**3+0.3*(math.sin(2*i*math.pi/250)+math.sin(2*i*math.pi/25))**2-0.4*(math.sin(2*i*math.pi/250)+math.sin(2*i*math.pi/25))
    print(u)
    uk.append(u)
```

```python
y=np.zeros((1,500))
y[0,1]=1.0
```

```python
for j in range(498):
    y[0,j+2]=0.3*y[0,j+1]+0.6*y[0,j]+uk[j]
print(y)
```

```python
from matplotlib import pyplot as plt

dev_x = range(0,500)
dev_y = y.squeeze(0)

plt.plot(dev_x, dev_y)
plt.show()
```

```python
ukk=[]
for i in x:
    uu=math.sin(2*i*math.pi/250)+math.sin(2*i*math.pi/25)
#     print(u)
    ukk.append(uu)
yy=np.zeros((1,500))
yy[0,1]=1.0
for j in range(498):
    yy[0,j+2]=0.3*yy[0,j+1]+0.6*yy[0,j]+ukk[j]
# print(yy)

from matplotlib import pyplot as plt

dx = range(0,500)
dy = yy.squeeze(0)

plt.plot(dx, dy)
plt.show()
```

#### We reproduce a difference discrete equation and we find that the more complex the function fitted, the better the ability to fit complex problems.

```python

import numpy as np
from matplotlib import pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']

W1=np.array([[0.15,0.20],[0.25,0.30]])
b1=0.35  
W2=np.array([[0.40,0.45],[0.50,0.55]])
b2=0.6
X=np.array([[0.05],[0.10]])
Y=np.array([[0.01],[0.99]])
alpha=0.5
Count=10000
m=2
 
def sigmoid(x):
    return 1/(1+np.exp(-x))
 
def NNtraining(W1,X,b1,W2,b2,Y,m,alpha):
    Count=1000
    J = np.zeros((Count, 1))
    for i in range(Count):
        Z1 = np.dot(W1, X) + b1
        A1 = sigmoid(Z1)
        Z2 = np.dot(W2, A1) + b2
        A2 = sigmoid(Z2)
        J[i]=-1/m*(np.dot(Y.T,np.log(A2)+np.dot((1-Y).T,np.log(1-A2))))
        # backpropagation algorithm
        dZ2=A2-Y
        dW2=1/m*np.dot(dZ2,A1.T)
        db2=1/m*np.sum(dZ2,axis=1,keepdims=True)
        dZ1=np.dot(W2.T,dZ2)*np.dot(Z1.T,1-Z1)
        dW1=1/m*np.dot(dZ1,X.T)
        db1=1/m*np.sum(dZ1,axis=1,keepdims=True)
        W1=W1-alpha*dW1
        b1=b1-alpha*db1
        W2=W1-alpha*dW2
        b2=b2-alpha*db2
#         print("result:",A2[0],A2[1])
    return J
 
if __name__ == '__main__':
    # W1,X,b1,W2,b2,Y,m,alpha=InitNN()
    J=NNtraining(W1,X,b1,W2,b2,Y,m,alpha)
    fig=plt.figure(1)
    plt.plot(J)
    plt.xlabel(u'iter')
    plt.ylabel(u'Loss')
    plt.show()
```
